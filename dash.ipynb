Agora veficando quantas dessas são escolas privadas em cada região

1) nordeste

```python
df_nordeste = df[df["NO_REGIAO"]=="Nordeste"]
nordeste_particular = df_nordeste[df_nordeste["TP_DEPENDENCIA"] != 4]
nordeste_publicas = df_nordeste[df_nordeste["TP_DEPENDENCIA"] == 4]
```

2) Sudeste

```python
df_sudeste = df[df["NO_REGIAO"]=="Sudeste"]
sudeste_particular = df_sudeste[df_sudeste["TP_DEPENDENCIA"] != 4]
sudeste_publicas = df_sudeste[df_sudeste["TP_DEPENDENCIA"] == 4]
```

3)Sul

```python
df_sul = df[df["NO_REGIAO"]=="Sul"]
sul_particular = df_sul[df_sul["TP_DEPENDENCIA"] != 4]
sul_publicas = df_sul[df_sul["TP_DEPENDENCIA"] == 4]
```

4)Norte

```python
df_norte = df[df["NO_REGIAO"]=="Norte"]
norte_particular = df_norte[df_norte["TP_DEPENDENCIA"] != 4]
norte_publicas = df_norte[df_norte["TP_DEPENDENCIA"] == 4]
```

5)Centro-Oeste

```python
df_CentroOeste = df[df["NO_REGIAO"]=="Centro-Oeste"]
CentroOeste_particular = df_CentroOeste[df_CentroOeste["TP_DEPENDENCIA"] != 4]
CentroOeste_publicas = df_CentroOeste[df_CentroOeste["TP_DEPENDENCIA"] == 4]
```

```python
hfreq = pd.DataFrame(series_v)
hfreq.rename(columns={'NO_REGIAO': 'QTD_ESCOLAS'},inplace=True)
hfreq["QTD_ESCOLAS_PARTICULARES"] = [len(nordeste_particular), len(sudeste_particular), len(sul_particular), len(norte_particular), len(CentroOeste_particular)]
hfreq["QTD_ESCOLAS_PUBLICAS"] = [len(nordeste_publicas), len(sudeste_publicas), len(sul_publicas), len(norte_publicas), len(CentroOeste_publicas)]
hfreq["FREQ_REL_ESCOLAS_PARTICULARES"] = hfreq["QTD_ESCOLAS_PARTICULARES"]/hfreq["QTD_ESCOLAS"]
hfreq
```

```python
app = Dash(__name__)

# assume you have a "long-form" data frame
# see https://plotly.com/python/px-arguments/ for more options

fig = px.bar(hfreq, x=hfreq.index,y="QTD_ESCOLAS_PARTICULARES",barmode="group")
app.layout = html.Div(children=[
     dcc.Dropdown(['QTD_ESCOLAS_PARTICULARES', 'QTD_ESCOLAS_PUBLICAS'], 'QTD_ESCOLAS_PUBLICAS', id='tipo_escola'),
    html.H1(children='Hello Dash'),

    html.Div(children='''
        Dash: A web application framework for your data.
    '''),

    dcc.Graph(
        id='escola_regioes',
        figure=fig
    )
])

@app.callback(
    Output('escola_regioes', 'figure'),
    Input('tipo_escola', 'value')
)
def update_output(value):
    fig = px.bar(hfreq, x=hfreq.index,y=value,barmode="group")
    return fig
if __name__ == '__main__':
    app.run_server()
```
